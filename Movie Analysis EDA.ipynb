```python
plt.figsize = (50,50)
sns.histplot(data ,y="IMDb Rating",x="Genre")
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Duration (minutes)', hue='Title', bins=10, kde=True, multiple="stack")

# Add labels
plt.xlabel('Duration (minutes)')
plt.ylabel('Title')

# Show the plot
plt.show()
```

```python
status_counts = data['Genre'].value_counts()
plt.figure(figsize=(10, 10))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140)
plt.show()
```

```python
sns.boxplot(x="Certificates", y="IMDb Rating", data=data)

plt.xlabel("Certificates")
plt.ylabel("IMDb Rating")
plt.show()
```

```python
sns.violinplot(x="Genre", y="Year", data=data)

plt.xlabel("Genre")
plt.ylabel("Year")
plt.show()
```

```python
corr_matrix = data.select_dtypes(include=['number']).corr()

# Create a heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(corr_matrix, annot=True, cmap='crest', linewidths=0.5)

plt.show()
```

```python
sns.pairplot(data.select_dtypes(include=['number']))

plt.show()
```
